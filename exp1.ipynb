importing the dependencies


```python
from sklearn.datasets import fetch_20newsgroups
from nltk import bigrams, trigrams
from random import choice
```

Data collection and data processing


```python
newsgroups_data = fetch_20newsgroups(subset='all', remove=('headers', 'footers', 'quotes'))
```

```python
text_data = [document.split() for document in newsgroups_data.data]
```

```python
bigram_model = list(bigrams(text_data))
trigram_model = list(trigrams(text_data))
```

```python
print(bigram_model[0])
```

```python
def generate_random_text(n, language_model):
    text = []
    for _ in range(n):
        word = choice(language_model)
        text.append(str(word[0]))  # Convert to string
        text.append(str(word[1]))  # Convert to string
    return ' '.join(text)

for i in range(12):
  random_text = generate_random_text(50, bigram_model)
  print(random_text)
```
